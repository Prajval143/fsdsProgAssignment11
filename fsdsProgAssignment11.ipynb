```python
def string_k(k, str):
    string = []
    text = str.split(" ")
    for x in text:
        if len(x) > k:
            string.append(x)
    return string

k = 3
str ="Python is good"
print(string_k(k, str))
```

```python
def remove(string, i): 
    a = string[ : i] 
    b = string[i + 1: ]
    return a + b
string = "PythonIsGood"
i = 5
print(remove(string, i))
```

```python
def check(string) :
    p = set(string)
    s = {'0', '1'}
    if s == p or p == {'0'} or p == {'1'}:
        print("Yes")
    else :
        print("No")
string = "101010000111"
check(string)
```

```python
def uncommon(a,b):
    a=a.split()
    b=b.split()
    k=set(a).symmetric_difference(set(b))
    return k


a="apple banana mango"
b="banana fruits mango"
print(list(uncommon(a,b)))
```

```python

string = "tutorialspoint"

duplicates = []
for char in string:
   
   if string.count(char) > 1:
       if char not in duplicates:
           duplicates.append(char)
print(*duplicates)
```
